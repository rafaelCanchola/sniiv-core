```python
import pandas as pd
tabla_principal = 'join.csv'
tabla_foranea = 'foreign.csv'
tabla_orden = 'columns.csv'
tabla_nombres = "tablas.txt"
tabla_ordenacion = 'ordenacion.txt'

tabla1 = pd.read_csv(tabla_principal)
foranea = pd.read_csv(tabla_foranea)
columnas = pd.read_csv(tabla_orden)
tablas = pd.read_csv(tabla_nombres)
ordenacion = pd.read_csv(tabla_ordenacion)

tabla2 = tabla1.merge(foranea, how="left", on=["NOM_CAMP","NOM_TABLA"], )
tabla3 = tabla2.merge(columnas[['NOM_CAMP','NOM_TABLA','ORDINAL_POSITION']],how="left",on=["NOM_CAMP","NOM_TABLA"], )
ordenacion.head(20)
```

```python
array = []
for column in ordenacion.itertuples():
    l = 'echo "'+column.TABLAS+': "\npsql -d sniiv_local --user developer -c "\copy '+column.TABLAS+' ('
    for fields in tabla3.itertuples():
        if fields.NOM_TABLA == column.TABLAS:
            l = l +fields.NOM_CAMP+','
    l = l[:-1] + ') FROM \'sniiv_'+column.TABLAS+'.csv\' DELIMITER \',\' ;"\n'
    array.append(l)
file = open("cargasDB.txt", "w+")
for row in array:
    file.write(row)
file.close()
```
