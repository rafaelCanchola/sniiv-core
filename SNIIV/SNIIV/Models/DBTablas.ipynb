```python
import pandas as pd
tabla_principal = 'join.csv'
tabla_foranea = 'foreign.csv'
tabla_primaria = 'primary.csv'
tabla_nombres = "tablas.txt"

tabla1 = pd.read_csv(tabla_principal)
foranea = pd.read_csv(tabla_foranea)
primaria = pd.read_csv(tabla_primaria)
tablas = pd.read_csv(tabla_nombres)
```

```python
tablas
```

```python
tabla2 = tabla1.merge(foranea, how="left", on=["NOM_CAMP","NOM_TABLA"], )
```

```python
tabla3 = tabla2.merge(primaria, how="left", on=["NOM_CAMP","NOM_TABLA"], )
```

```python
tabla3[tabla3['pk_name']=="PRIMARY"]
```

```python
tabla3
```

```python
tabla3['CLAVE'] = tabla3.apply(lambda x: "Interna" if (x['CLAVE']=='Foranea' or x['CLAVE']=='foranea' or x['CLAVE']=='interna' or x['CLAVE']=='principal' or x['CLAVE']=='Principal') else x['CLAVE'], axis=1)
```

```python
tabla3['CLAVE'].value_counts()
```

```python
tabla3['CLAVE'] = tabla3.apply(lambda x: "Primaria" if x['pk_name']=="PRIMARY" else x['CLAVE'],axis=1)
```

```python
tabla3['CLAVE'].value_counts()
```

```python
tabla3['CLAVE'] = tabla3.apply(lambda x: "Foranea" if pd.notna(x['FOR_TABLA'])  else x['CLAVE'],axis=1)
```

```python
tabla3['CLAVE'].value_counts()
```

```python
tabla3[tabla3["FOR_TABLA"].notna()]
```

```python
tabla3[tabla3["FOR_TABLA"].notna() & tabla3["pk_name"].notna()]
```

```python
tabla3["TIPO"].value_counts()
```

```python
tabla3[tabla3["TIPO"]=='DATE']
```

```python
tabla3['TIPO'] = tabla3.apply(lambda x: 'VARCHAR' if (x['TIPO']=='ENUM' or x['TIPO']=='CHAR') else x['TIPO'],axis=1)
```

```python
tabla3['TIPO'] = tabla3.apply(lambda x: 'INT' if (x['TIPO']=='YEAR' or x['TIPO']=='TINYINT' or x['TIPO']=='SMALLINT' or x['TIPO']=='MEDIUMINT' or x['TIPO']=='BIGINT') else x['TIPO'],axis=1)
```

```python
tabla3['TIPO'] = tabla3.apply(lambda x: 'DECIMAL' if x['TIPO']=='FLOAT' else x['TIPO'],axis=1)
```

```python
tabla3['TIPO'] = tabla3.apply(lambda x: 'DATE' if (x['TIPO']=='FECHA') else x['TIPO'],axis=1)
```

```python
tabla3["TIPO"].value_counts()
```

```python
tabla3[(tabla3["TAMAÑO"].isna())&(tabla3["TIPO"]=="VARCHAR")]
```

```python
tabla3['TAMAÑO'] = tabla3.apply(lambda x: '10' if (x['TIPO'] == 'VARCHAR') and ( x['TAMAÑO'] ) else x['TAMAÑO'],axis=1)
```

```python
tabla3[(tabla3["TAMAÑO"].isna())&(tabla3["TIPO"]=="VARCHAR")]
```

```python
tabla3['MODELADO'] = tabla3.apply(lambda x: '[Key] ' if x['CLAVE']=='Primaria' else '',axis=1) + tabla3.apply(lambda x: '[StringLength(' + str(x['TAMAÑO']) + ')] ' if x['TIPO']=='VARCHAR' else '',axis=1) + 'public ' + tabla3.apply(lambda x: 'int' if x['TIPO']=='INT' else ('string' if x['TIPO']=='VARCHAR' else ('Decimal' if x["TIPO"] == "DECIMAL" else ('DateTime' if x["TIPO"]=="DATE" else ('Boolean' if x["TIPO"]=="BIT" else '')))),axis=1)+' '+ tabla3['NOM_CAMP'] + " {get; set;}" +tabla3.apply(lambda x: "\n		public " + x["FOR_TABLA"] +" "+ x["FOR_TABLA"] +" {get; set;}" if x["CLAVE"]=="Foranea" else '',axis=1)
```

```python
tabla3["MODELADO"].sample(10)
```

```python
tabla_final = tabla3[['NOM_TABLA', 'MODELADO']].copy()
```

```python
tabla_final
```

```python
tabla_final.to_csv('tabla_final.csv',index = False,columns=['NOM_TABLA', 'MODELADO'])
```
